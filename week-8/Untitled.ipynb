```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSVs you showed in your folder
dx = pd.read_csv("iPhone Thing-Accelerometer_X.csv", parse_dates=["time"])
dy = pd.read_csv("iPhone Thing-Accelerometer_Y.csv", parse_dates=["time"])
dz = pd.read_csv("iPhone Thing-Accelerometer_Z.csv", parse_dates=["time"])

# Normalise column names
for df in (dx, dy, dz):
    df.columns = [c.strip().lower() for c in df.columns]  # usually ['time','value']

# 1) Plot X
plt.figure()
plt.plot(dx["time"], dx["value"])
plt.title("Accelerometer X")
plt.xlabel("Time"); plt.ylabel("m/s")
plt.tight_layout()

# 2) Plot Y
plt.figure()
plt.plot(dy["time"], dy["value"])
plt.title("Accelerometer Y")
plt.xlabel("Time"); plt.ylabel("m/s")
plt.tight_layout()

# 3) Plot Z
plt.figure()
plt.plot(dz["time"], dz["value"])
plt.title("Accelerometer Z")
plt.xlabel("Time"); plt.ylabel("m/s")
plt.tight_layout()

# 4) Plot all together
plt.figure()
plt.plot(dx["time"], dx["value"], label="X")
plt.plot(dy["time"], dy["value"], label="Y")
plt.plot(dz["time"], dz["value"], label="Z")
plt.legend()
plt.title("Accelerometer X/Y/Z")
plt.xlabel("Time"); plt.ylabel("m/s")
plt.tight_layout()
```
